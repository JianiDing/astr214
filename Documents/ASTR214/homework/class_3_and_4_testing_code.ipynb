```python
from colossus.cosmology import cosmology
import numpy as np
from matplotlib import pylab as plt
from classcode.plot_utils import plot_pretty
plot_pretty(dpi=300)
```

```python
my_cosmo_eds = {'flat': True, 'H0': 73.0, 'Om0': 0.8, 'Ob0': 0.045714, 'sigma8': 0.74, 'ns': 0.951}
my_cosmo_lcdm = {'flat': True, 'H0': 73.0, 'Om0': 0.238, 'Ob0': 0.045714, 'sigma8': 0.74, 'ns': 0.951}
my_cosmo_lcdm2 = {'flat': True, 'H0': 73.0, 'Om0': 0.6, 'Ob0': 0.045714, 'sigma8': 0.74, 'ns': 0.951}
my_cosmo_lcdm3 = {'flat': True, 'H0': 73.0, 'Om0': 0.4, 'Ob0': 0.045714, 'sigma8': 0.74, 'ns': 0.951}
# set my_cosmo to be the current cosmology
cosmo = cosmology.setCosmology('my_cosmo', my_cosmo_lcdm)
cosmo2 = cosmology.setCosmology('eds', my_cosmo_eds)
cosmo3 = cosmology.setCosmology('my_cosmo', my_cosmo_lcdm2)
cosmo4 = cosmology.setCosmology('my_cosmo', my_cosmo_lcdm3)
```

```python
z=np.arange(0.,200.,0.02) 
plt.xscale('log')
plt.plot(z,cosmo.Om(z),c='slateblue',
         linewidth=2., label='$\Omega_{\rm}$')
plt.plot(z,cosmo.Ogamma(z),c='red',
         linewidth=2., label='$\Omega_{{\Gamma}}$')
plt.plot(z,cosmo.OL(z),c='grey',
         linewidth=2., label='$\Omega_{{\Lambda}}$')
plt.plot(z,cosmo.OL(z)+cosmo.Om(z)+cosmo.Ogamma(z)+cosmo.Onu(z),c='green',
         linewidth=2., label='$\Omega_{{\total}}$')
plt.legend()
plt.show()
```

```python
plt.xscale('log')
plt.yscale('log')
plt.plot(z, cosmo.growthFactor(z), c='blue', linewidth=2., 
         label= r'Growth Factor $(z < 200)$ $\Lambda CDM$')
#axvline(x=linewidth=4)
plt.plot(z, cosmo2.growthFactor(z), c='red', linewidth=2., 
         label= r'Growth Factor $\Omega_{m} = 0.8 \  z < 200$ EDS')
plt.plot(z, cosmo3.growthFactor(z), c='green', linewidth=2., 
         label= r'Growth Factor $\Omega_{m} = 0.6 \ z < 200$ $\Lambda CDM$')
plt.plot(z, cosmo4.growthFactor(z), c='k', linewidth=2., 
         label= r'Growth Factor $\Omega_{m} = 0.4 \ z < 200$ $\Lambda CDM$')
plt.xlabel('z')
plt.ylabel('Growth Factor')

plt.legend()
plt.show()
```
